```python
Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set default style
sns.set(style="whitegrid")
```

```python
 # Loading dataset
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
df = pd.read_csv(url)

# Preview the data
df.head()

```

# ***Visualization***

1. Count of Each `Species


```python
sns.countplot(x='species', data=df)
plt.title("Count of Each Iris Species")
plt.xlabel("Species")
plt.ylabel("Count")
plt.show()
```

2. Sepal Length vs Sepal  width

```python
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=df)
plt.title("Sepal Length vs Sepal Width by Species")
plt.show()
```

3. Petal Length vs Petal Width

```python
sns.scatterplot(x='petal_length', y='petal_width', hue='species', style='species', data=df)
plt.title("Petal Length vs Petal Width by Species")
plt.show()
```

4. Distribution of Petal Length

```python
sns.histplot(df['petal_length'], kde=True, bins=20)
plt.title("Distribution of Petal Length")
plt.xlabel("Petal Length (cm)")
plt.show()
```

 5. Boxplot: Sepal Length by Species

```python
sns.boxplot(x='species', y='sepal_length', data=df)
plt.title("Sepal Length Distribution by Species")
plt.show()
```

6. Pairplot (Bonus – All Features Together)

```python
sns.pairplot(df, hue='species')
plt.suptitle("Pairwise Feature Comparison", y=1.02)
plt.show()
```
